–í—ã–≤–µ–¥–∏—Ç–µ ùë° —Å—Ç—Ä–æ–∫. –û—á–µ—Ä–µ–¥–Ω–∞—è —Å—Ç—Ä–æ–∫–∞ –¥–æ–ª–∂–Ω–∞ —Å–æ–¥–µ—Ä–∂–∞—Ç—å -1, –µ—Å–ª–∏ –¥–ª—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–≥–æ –Ω–∞–±–æ—Ä–∞ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –∏—Å–∫–æ–º–æ–π –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç. –í –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ –≤—ã–≤–µ–¥–∏—Ç–µ –∏—Å–∫–æ–º—É—é –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫—É. –ï—Å–ª–∏ —Ç–∞–∫–∏—Ö –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ, —Ç–æ –≤—ã–≤–µ–¥–∏—Ç–µ –ª—é–±—É—é –∏–∑ –Ω–∏—Ö.

```python
def algorithm(n):
    lis = []
    
    if (n < 4): #—Ç.–∫. –≤ —Ç–∞–∫–æ–º —Å–ª—É—á–∞–µ –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ —Å–æ–±–ª—é—Å—Ç–∏ —Ä–∞–∑—Ä—ã–≤ –º–µ–∂–¥—É —á–∏—Å–ª–∞–º–∏ –≤ –¥–∏–∞–ø–æ–∑–æ–Ω–µ –æ—Ç 2 –¥–æ 4, –∞ –ø—Ä–∏ n = 4 –∏ –±–æ–ª—å—à–µ - –º–æ–∂–Ω–æ
        return -1
    
    if n%2 != 0:
        for i in range(n,0,-2): #–≤—ã–≤–æ–∂—É –≤—Å–µ –Ω–µ—á—ë—Ç–Ω—ã–µ —á–∏—Å–ª–∞ –≤ –ø–æ—Ä—è–¥–∫–µ —É–±—ã–≤–∞–Ω–∏—è
            lis.append(i)
        
        lis.append(4) #"–∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π" –º–æ–º–µ–Ω—Ç, —Ç.–∫. –∏–Ω–∞—á–µ –∏–¥—ë—Ç 1, 2 - –Ω–µ —Å–æ–±–ª—é–¥–∞–µ—Ç—Å—è —É—Å–ª–æ–≤–∏–µ
        lis.append(2)

        for i in range(6,n-1,2):  #–≤—ã–≤–æ–∂—É —á—ë—Ç–Ω—ã–µ —á–∏—Å–ª–∞ –≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è, –Ω–∞—á–∏–Ω–∞—é —Å 6, —Ç.–∫. 2 –∏ 4 —É–∂–µ –Ω–∞–ø—á–µ—Ç–∞–Ω–∞–Ω—ã
            lis.append(i)
        return lis
    else:
        for i in range(n-1,0,-2): #–≤—ã–≤–æ–∂—É –≤—Å–µ –Ω–µ—á—ë—Ç–Ω—ã–µ —á–∏—Å–ª–∞ –≤ –ø–æ—Ä—è–¥–∫–µ —É–±—ã–≤–∞–Ω–∏—è
            lis.append(i)
        
        lis.append(4)
        lis.append(2)

        for i in range(6,n-1,2):  #–≤—ã–≤–æ–∂—É —á—ë—Ç–Ω—ã–µ —á–∏—Å–ª–∞ –≤ –ø–æ—Ä—è–¥–∫–µ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—è
            lis.append(i)
        lis.append(n) #–ø–æ—Å–∫–æ–ª—å–∫—É –Ω–∞—á–∞–ª–∞ —Å –Ω–µ—á–µ—Ç–Ω–æ–≥–æ, —Ç–æ –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ–µ —á—ë—Ç–Ω–æ–µ —á–∏—Å–ª–æ "–≤–∏—Å–∏—Ç", –Ω—É–∂–Ω–æ –µ–≥–æ –¥–æ–±–∞–≤–∏—Ç—å, –æ–Ω–æ –±—É–¥–µ—Ç –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º
                        #–∏ —á—ë—Ç–Ω—ã–º, –ø–æ—ç—Ç–æ–º—É –≤ –∫–æ–Ω–µ—Ü –∑–∞–ø–∏—Å—ã–≤–∞—é
        return lis

    
t = int(input('–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª-–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: '))

if t < 1 or t > 100: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
        raise Exception('t –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å >= 1 –∏ t <= 100')
    
arr = []

for i in range(0,t):
    n = int(input('–í–≤–µ–¥–∏—Ç–µ –æ–ø–∏—Å–∞–Ω–∏–µ –Ω–∞–±–æ—Ä–∞ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö (n): '))
    
    if n < 2 or n > 1000: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
        raise Exception('n –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å >= 2 –∏ n <= 1000')
        
    arr.append(n)

print("----------------------------------------------")

for i in arr:
    result = algorithm(i)
    print(*result) #–≤—ã–≤–æ–¥–∏—Ç –∫—Ä–∞—Å–∏–≤–æ –º–∞—Å—Å–∏–≤
```

```python
def null_algorithm(l,r):
    
    if l > r: #–µ—Å–ª–∏ –∏–Ω–¥–µ–∫—Å –Ω–∞—á–∞–ª–∞ –±–æ–ª—å—à–µ –∏–Ω–¥–µ–∫—Å–∞ –∫–æ–Ω—Ü–∞, —Ç–æ —ç—Ç–æ –Ω–µ –∏–º–µ–µ—Ç —Å–º—ã—Å–ª–∞
        return
    
    e = (l+r)//2 #—Ñ–æ—Ä–º—É–ª–∞ –¥–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –∏–Ω–µ–¥–∫—Å–∞ –¥–ª—è –Ω–µ—á–µ—Ç–Ω–æ–≥–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ r-l+1
    a[e] = (l-r,e) #–≤ –º–∞—Å—Å–∏–≤ –∞ –∫–ª–∞–¥–µ—Ç—Å—è —Å–µ—Ç –∏–∑ –¥–≤—É—Ö –∑–Ω–∞—á–µ–Ω–∏–π
    null_algorithm(l,e-1) #—Ä–µ–∫—É—Ä—Å–∏—è –¥–ª—è –ª–µ–≤–æ–≥–æ –ø–æ–¥–º–∞—Å—Å–∏–≤–∞; –≤—Å–µ–≥–¥–∞ –±—É–¥–µ—Ç –¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –¥–≤–∞ –ø–æ–¥–º–∞—Å—Å–∏–≤–∞ —Å –∫–∞–∂–¥–æ–π —Å—Ç–æ—Ä–æ–Ω—ã
    null_algorithm(e+1,r) #—Ä–µ–∫—É—Ä—Å–∏—è –¥–ª—è –ø—Ä–∞–≤–æ–≥–æ –ø–æ–¥–º–∞—Å—Å–∏–≤–∞

t = int(input('–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª-–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –¥–∞–Ω–Ω—ã—Ö: '))

if t < 1 or t > 10**4: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
        raise Exception('t –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å >= 1 –∏ t <= 10**4')
        
arr = [] #–º–∞—Å—Å–∏–≤ –¥–ª—è –∑–Ω–∞—á–µ–Ω–∏–π n
sum = 0

for i in range(t):
    n = int(input('–í–≤–µ–¥–∏—Ç–µ –¥–ª–∏–Ω–Ω—É –∞:')) + 1 #—Ç.–∫. –æ—Ç—Ä–µ–∑–∫–∏ —Å—á–∏—Ç–∞—é—Ç—Å—è —Å 1, –∞ –Ω—É–º–µ—Ä–∞—Ü–∏—è –º–∞—Å—Å–∏–≤–∞ —Å 0 (n - –¥–ª–∏–Ω–∞ –º–∞—Å—Å–∏–≤–∞ –∏–∑ 0)
    
    if n < 1 or n > 210: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
        raise Exception('n –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å >= 1 –∏ n <= 2*10**5')
        
    arr.append(n)
    sum += n
    
    if sum > 210: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
        raise Exception('–°—É–º–º–∞ –Ω–∞–±–æ—Ä–æ–≤ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö –Ω–µ –¥–æ–ª–∂–Ω–∞ –ø—Ä–µ–≤—ã—à–∞—Ç—å 2*10**5!')

print("----------------------------------------------")

a = b = [0]*n #–∑–∞–ø–æ–ª–Ω—è—é –¥–≤–∞ –º–∞—Å—Å–∏–≤–∞ –Ω—É–ª—è–º–∏

for z in arr:
    
    null_algorithm(1,z-1)
    for i,j in enumerate(sorted(a[1:z])): #—Å–æ—Ä—Ç–∏—Ä—É—é—Ç—Å—è —Å–µ—Ç—ã –ø–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–Ω–∏—é (–≤ —Ç–∞–∫–æ–º –ø–æ—Ä—è–¥–∫–µ –∑–∞–ø–æ–ª–Ω—è—é—Ç—Å—è –¥–µ–π—Å—Ç–≤—Ç–∏—è i+1)
                                          #–Ω—É–ª–µ–≤–æ–µ —á–∏—Å–ª–æ l-r, –ø–µ—Ä–≤–æ–µ –∏–Ω–¥–µ–∫—Å, –ø–æ –∫–æ—Ç–æ—Ä–æ–º—É –∑–∞–ø–æ–ª–Ω—è–µ—Ç—Å—è, —Ç–æ–≥–¥–∞ j[1] —ç—Ç–æ –∏–Ω–¥–µ–∫—Å
        b[j[1]] = i+1 #b - –º–∞—Å—Å–∏–≤ —Å 0, –∑–∞–º–µ–Ω—è—é –≤ –Ω—ë–º –ø–æ –∏–Ω–¥–µ–∫—Å–∞–º –Ω–∞ –Ω—É–∂–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
    print(*b[1:z]) #–∫—Ä–∞—Å–∏–≤–æ –≤—ã–≤–æ–∂—É –º–∞—Å—Å–∏–≤
```

–ù–∞ –∫—É—Ö–Ω–µ —É –ü–æ–ª–∏–∫–∞—Ä–ø–∞ –≤ –Ω–∞–ª–∏—á–∏–∏ nb –∫—É—Å–æ—á–∫–æ–≤ —Ö–ª–µ–±–∞, ns –∫—É—Å–æ—á–∫–æ–≤ –∫–æ–ª–±–∞—Å—ã –∏ nc –∫—É—Å–æ—á–∫–æ–≤ —Å—ã—Ä–∞. –ö—Ä–æ–º–µ —Ç–æ–≥–æ, –≤ –º–∞–≥–∞–∑–∏–Ω–µ –Ω–µ–ø–æ–¥–∞–ª–µ–∫—É –µ—Å—Ç—å –≤ –ø—Ä–æ–¥–∞–∂–µ –≤—Å–µ —Ç—Ä–∏ –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–∞ –ø–æ —Ü–µ–Ω–µ: pb —Ä—É–±–ª–µ–π –∑–∞ –∫—É—Å–æ–∫ —Ö–ª–µ–±–∞, ps ‚Äî –∑–∞ –∫—É—Å–æ–∫ –∫–æ–ª–±–∞—Å—ã –∏ pc ‚Äî –∑–∞ –∫—É—Å–æ—á–µ–∫ —Å—ã—Ä–∞.

–£ –ü–æ–ª–∏–∫–∞—Ä–ø–∞ –µ—Å—Ç—å r —Ä—É–±–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –æ–Ω –≥–æ—Ç–æ–≤ –ø–æ—Ç—Ä–∞—Ç–∏—Ç—å –≤ –º–∞–≥–∞–∑–∏–Ω–µ. –ö–∞–∫–æ–µ –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–∞–º–±—É—Ä–≥–µ—Ä–æ–≤ –æ–Ω —Å–º–æ–∂–µ—Ç –ø—Ä–∏–≥–æ—Ç–æ–≤–∏—Ç—å? –°—á–∏—Ç–∞–µ—Ç—Å—è, —á—Ç–æ –ª–æ–º–∞—Ç—å –∏–ª–∏ —Ä–µ–∑–∞—Ç—å –ª—é–±–æ–π –∏–∑ –∫—É—Å–æ—á–∫–æ–≤ (—Ö–ª–µ–±–∞, –∫–æ–ª–±–∞—Å—ã, —Å—ã—Ä–∞) –Ω–µ–ª—å–∑—è, –∞ —Ç–∞–∫–∂–µ —á—Ç–æ –≤ –º–∞–≥–∞–∑–∏–Ω–µ –µ—Å—Ç—å –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ –º–Ω–æ–≥–æ –∫—É—Å–æ—á–∫–æ–≤ –∫–∞–∂–¥–æ–≥–æ –∏–∑ –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤.

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö —Å–æ–¥–µ—Ä–∂–∏—Ç—Å—è –Ω–µ–ø—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞, –æ–ø–∏—Å—ã–≤–∞—é—â–∞—è —Ä–µ—Ü–µ–ø—Ç ¬´–≥–∞–º–±—É—Ä–≥–µ—Ä–∞ –æ—Ç –ü–æ–ª–∏–∫–∞—Ä–ø–∞¬ª. –î–ª–∏–Ω–∞ —Å—Ç—Ä–æ–∫–∏ –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç 100, —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç —Ç–æ–ª—å–∫–æ –±—É–∫–≤—ã 'B' (–ø—Ä–æ–ø–∏—Å–Ω–∞—è –ª–∞—Ç–∏–Ω—Å–∫–∞—è B), 'S' (–ø—Ä–æ–ø–∏—Å–Ω–∞—è –ª–∞—Ç–∏–Ω—Å–∫–∞—è S) –∏ 'C' (–ø—Ä–æ–ø–∏—Å–Ω–∞—è –ª–∞—Ç–∏–Ω—Å–∫–∞—è C).

–í—Ç–æ—Ä–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç —Ç—Ä–∏ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ nb, ns, nc (1‚Äâ‚â§‚Äânb,‚Äâns,‚Äânc‚Äâ‚â§‚Äâ100) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫—É—Å–æ—á–∫–æ–≤ —Ö–ª–µ–±–∞, –∫–æ–ª–±–∞—Å—ã –∏ —Å—ã—Ä–∞ –Ω–∞ –∫—É—Ö–Ω–µ –ü–æ–ª–∏–∫–∞—Ä–ø–∞. –¢—Ä–µ—Ç—å—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç —Ç—Ä–∏ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ pb, ps, pc (1‚Äâ‚â§‚Äâpb,‚Äâps,‚Äâpc‚Äâ‚â§‚Äâ100) ‚Äî —Ü–µ–Ω–∞ –æ–¥–Ω–æ–≥–æ –∫—É—Å–æ—á–∫–∞ —Ö–ª–µ–±–∞, –∫–æ–ª–±–∞—Å—ã –∏ —Å—ã—Ä–∞ –≤ –º–∞–≥–∞–∑–∏–Ω–µ. –ù–∞–∫–æ–Ω–µ—Ü, —á–µ—Ç–≤–µ—Ä—Ç–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ r (1‚Äâ‚â§‚Äâr‚Äâ‚â§‚Äâ10**12) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä—É–±–ª–µ–π —É –ü–æ–ª–∏–∫–∞—Ä–ø–∞.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
–í—ã–≤–µ–¥–∏—Ç–µ –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–∞–º–±—É—Ä–≥–µ—Ä–æ–≤, –∫–æ—Ç–æ—Ä–æ–µ –ü–æ–ª–∏–∫–∞—Ä–ø –º–æ–∂–µ—Ç –ø—Ä–∏–≥–æ—Ç–æ–≤–∏—Ç—å. –ï—Å–ª–∏ –æ–Ω –Ω–µ –º–æ–∂–µ—Ç –ø—Ä–∏–≥–æ—Ç–æ–≤–∏—Ç—å –Ω–∏ –æ–¥–Ω–æ–≥–æ –≥–∞–º–±—É—Ä–≥–µ—Ä–∞, –≤—ã–≤–µ–¥–∏—Ç–µ 0.

```python
def reciept(l, r, n, p, rub): #r - —á–∏—Å–ª–æ –±–æ–ª—å—à–µ–µ –º–∞–∫—Å–∏–º—É–º–∞ —Ä—É–±–ª–µ–π –ü–æ–ª–∏–∫–∞—Ä–ø–∞, n - –∫–æ–ª-–≤–æ –∫—É—Å–æ—á–∫–æ–≤, p - —Ü–µ–Ω–∞ –∫—É—Å–æ—á–∫–æ–≤,
                              #rub - –¥–µ–Ω—å–≥–∏ –ü–æ–ª–∏–∫–∞—Ä–ø–∞
    
    while l + 1 < r:
    
        mid = (l+r)//2 #–±–µ—Ä—ë—Ç—Å—è —Å–µ—Ä–µ–¥–∏–Ω–∞
        c = 0
        for i in range(3):
            c += p[i]*max(a[i]*mid - n[i],0) #—Å–∫–ª–∞–¥—ã–≤–∞–µ—Ç –¥–ª—è –∫–∞–∂–¥–æ–π –±—É–∫–≤—ã —Å—É–º–º—É, –Ω–∞—á–∏–Ω–∞–µ–º
                                             #—Å –º–∞–∫—Å–∏–º—É–º—É–∞ –¥–µ–Ω–µ–≥ —É –ü–æ–ª–∏–∫–∞—Ä–ø–∞, —Ñ—É–Ω–∫—Ü–∏—è –º–∞–∫—Å–∏–º—É–º–∞ —Ä–∞–±–æ—Ç–∞–µ—Ç –¥–ª—è —Å–ª—É—á–∞–µ–≤ –ø–æ–ª—É—á–µ–Ω–∏—è –æ—Ç—Ä–∏—Ü.
                                             #–≤–µ–ª–∏—á–∏–Ω; –≤ –º–∞—Å—Å–∏–≤–µ –∞ —Å–æ–¥–µ—Ä–∂–∏—Ç—Å—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–∞–∂–¥—ã—Ö –±—É–∫–≤
        if c > rub:
            r = mid
        else:
            l = mid
        
    return l

s = list(input('–í–≤–µ–¥–∏—Ç–µ —Ä–µ—Ü–µ–ø—Ç: '))

if len(s) < 1 or len(s) > 100: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
        raise Exception('–î–ª–∏–Ω–∞ —Å—Ç—Ä–æ–∫–∏ –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç 100 –∏ –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –º–µ–Ω—å—à–µ 1 —Å–∏–º–≤–æ–ª–∞')
for i in s: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
    if i != 'B' and i != 'C' and i != 'S':
        raise Exception('–í–≤–µ–¥—ë–Ω–Ω–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç –Ω–µ–¥–æ–ø—É—Å—Ç–∏–º—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è!')
        
a = [s.count(c) for c in 'BSC'] #—Å—á–∏—Ç–∞—é, —Å–∫–æ–ª—å–∫–æ –∫—É—Å–æ—á–∫–æ–≤ —Ö–ª–µ–±–∞, —Å—ã—Ä–∞ –∏ –∫–æ–ª–±–∞—Å—ã
n = list(map(int,input('–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª-–≤–æ —Ö–ª–µ–±–∞, –∫–æ–ª–±–∞—Å—ã –∏ —Å—ã—Ä–∞ –Ω–∞ –∫—É—Ö–Ω–µ: ').split()))

if len(n) != 3: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
    raise Exception('–ù—É–∂–Ω–æ –≤–≤–µ—Å—Ç–∏ 3 —á–∏—Å–ª–∞!')
for i in n:
    if i > 100 or i < 1:
        raise Exception('–ö–æ–ª-–≤–æ –∫—É—Å–æ—á–∫–æ–≤ –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –±–æ–ª—å—à–µ 100 –∏–ª–∏ 0!')
        
p = list(map(int,input('–í–≤–µ–¥–∏—Ç–µ —Ü–µ–Ω—É –∫—É—Å–æ—á–∫–∞ —Ö–ª–µ–±–∞, –∫–æ–ª–±–∞—Å—ã –∏ —Å—ã—Ä–∞ –≤ –º–∞–≥–∞–∑–∏–Ω–µ: ').split()))
    
if len(p) != 3: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
    raise Exception('–ù—É–∂–Ω–æ –≤–≤–µ—Å—Ç–∏ 3 –∑–Ω–∞—á–µ–Ω–∏—è —Ü–µ–Ω—ã!')
for i in p:
    if i > 100 or i < 1:
        raise Exception('–¶–µ–Ω–∞ –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –±–æ–ª—å—à–µ 100 –∏–ª–∏ —Ä–∞–≤–Ω–∞ 0!')
        
rub = int(input('–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª-–≤–æ —Ä—É–±–ª–µ–π —É –ü–æ–ª–∏–∫–∞—Ä–ø–∞: '))

if rub > 10**12 or rub < 1:
    raise Exception('–ü–æ–ª–∏–∫–∞—Ä–ø —Ä–∞–±–æ—Ç–∞–µ—Ç, —É –Ω–µ–≥–æ –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å 0 —Ä—É–±–ª–µ–π, –Ω–æ –∏ –æ–Ω –Ω–µ –±–æ–≥–∞—á, —É –Ω–µ–≥–æ –Ω–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –±–æ–ª—å—à–µ 10**12 —Ä—É–±–ª–µ–π!')

print('–ù–∞–∏–±–æ–ª—å—à–µ–µ –∫–æ–ª-–≤–æ –≥–∞–º–±—É—Ä–≥–µ—Ä–æ–≤ = ', reciept(0, 10**13, n, p, rub))
```

–ü–µ—Ä–≤–∞—è —Å—Ç—Ä–æ–∫–∞ –∫–∞–∂–¥–æ–≥–æ –Ω–∞–±–æ—Ä–∞ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö —Å–æ–¥–µ—Ä–∂–∏—Ç –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ ùëõ –∏ ùë• (1‚â§ùëõ‚â§1000,1‚â§ùë•‚â§ùëõ) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–µ—Ä—à–∏–Ω –≤ –¥–µ—Ä–µ–≤–µ –∏ —Å–ø–µ—Ü–∏–∞–ª—å–Ω—É—é –≤–µ—Ä—à–∏–Ω—É, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ.

–ö–∞–∂–¥–∞—è –∏–∑ —Å–ª–µ–¥—É—é—â–∏—Ö ùëõ‚àí1 —Å—Ç—Ä–æ–∫ —Å–æ–¥–µ—Ä–∂–∏—Ç –¥–≤–∞ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ ùë¢, ùë£ (1‚â§ùë¢,ùë£‚â§ùëõ, ùë¢‚â†ùë£), —á—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –º–µ–∂–¥—É –≤–µ—Ä—à–∏–Ω–∞–º–∏ ùë¢ –∏ ùë£ –µ—Å—Ç—å —Ä–µ–±—Ä–æ.

–í –ø–µ—Ä–≤–æ–º –Ω–∞–±–æ—Ä–µ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö Ayush –º–æ–∂–µ—Ç —É–¥–∞–ª–∏—Ç—å —Ç–æ–ª—å–∫–æ –≤–µ—Ä—à–∏–Ω—É 2 –∏–ª–∏ 3, –ø–æ—Å–ª–µ —á–µ–≥–æ –≤–µ—Ä—à–∏–Ω–∞ 1 —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –ª–∏—Å—Ç–æ–º, –∏ Ashish –º–æ–∂–µ—Ç —É–¥–∞–ª–∏—Ç—å –µ–µ –≤ —Å–≤–æ—é –æ—á–µ—Ä–µ–¥—å.
–í–æ –≤—Ç–æ—Ä–æ–º –Ω–∞–±–æ—Ä–µ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö Ayush –º–æ–∂–µ—Ç —É–¥–∞–ª–∏—Ç—å –≤–µ—Ä—à–∏–Ω—É 2 –Ω–∞ —Å–∞–º–æ–º –ø–µ—Ä–≤–æ–º —à–∞–≥–µ.

```python
def derevo(t, n, vershu, vershv):
    
    while t > 0:
        
        degree = [0]*(n+1) #–±—É–¥–µ—Ç n —ç–ª–µ–º–µ–Ω—Ç–æ–≤, –ø—Ä–æ—Å—Ç–æ –Ω—É–º–µ—Ä–∞—Ü–∏—è —Å 0, –∞ —è –∏—Å–ø–æ–ª—å–∑—É—é –æ—Ç—Ä–µ–∑–æ–∫ –æ—Ç 1 –¥–æ n+1
                           #–∫–ª–∞–¥—É –≤ —Ç–∞–±–ª–∏—Ü—É –ø–æ–¥ –Ω–æ–º–µ—Ä–æ–º –≤–µ—Ä—à–∏–Ω—ã (–∏–Ω–¥–µ–∫—Å) –µ–¥–∏–Ω–∏—Ü—É, —Ç.–µ. –µ—Å–ª–∏ –º–µ–∂–¥—É 2 –∏ 3 –≤–µ—Ä—à–∏–Ω–æ–π –µ—Å—Ç—å –ø—É—Ç—å
                           #—Ç–æ degree[2] = 1 –∏ degree[3] = 1; –µ—Å–ª–∏ –≤ –æ–¥–Ω–æ–π –≤–µ—Ä—à–∏–Ω–µ –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø—É—Ç–µ–π, —Ç–æ –±–æ–ª—å—à–µ –µ–¥–∏–Ω–∏—Ü—ã –±—É–¥–µ—Ç
                           #—Ç–∞–∫ –ø–æ–ª—É—á–∞–µ—Ç—Å—è —É—Å–ª–æ–≤–∏–µ –¥–æ—Å—Ç–∏–∂–µ–Ω–∏—è —Å–ø–µ—Ü–∏–∞–ª—å–Ω–æ–π –≤–µ—Ä—à–∏–Ω—ã <= 1
                    
        for i in range(n-1): #–º–µ–∂–¥—É –≤–µ—Ä—à–∏–Ω–∞–º–∏ u, v –µ—Å—Ç—å —Ä–µ–±—Ä–æ
                
            degree[vershu[i]] += 1
            degree[vershv[i]] += 1
        
        if degree[x] <= 1:
            return 'Ayush'
        elif n%2:
            return 'Ashish'
        else:
            return 'Ayush'
        
        t -= 1

t = int(input('–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª-–≤–æ –Ω–∞–±–æ—Ä–æ–≤ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö: '))

if t < 1 or t > 10: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
    raise Exception('t –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å >= 1 –∏ t <= 10')

n, x = map(int, input().split()) #–∫–æ–ª-–≤–æ –≤–µ—Ä—à–∏–Ω –≤ –¥–µ—Ä–µ–≤–µ –∏ —Å–ø–µ—Ü–∏–∞–ª—å–Ω–∞—è –≤–µ—Ä—à–∏–Ω–∞
vershu = [0]*n
vershv = [0]*n

if n < 1 or n > 1000: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
    raise Exception('n –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å >= 1 –∏ n <= 1000')
if x < 1 or x > n: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
    raise Exception('x –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å >= 1 –∏ x <= n')
    
for i in range(n-1): #–º–µ–∂–¥—É –≤–µ—Ä—à–∏–Ω–∞–º–∏ u, v –µ—Å—Ç—å —Ä–µ–±—Ä–æ
    u, v = map(int, input().split())
     
    if u < 1 or u == v: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
        raise Exception('u –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å >= 1 –∏ u != v')
    if v > n: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
        raise Exception('v –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å <= n')
                
    vershu[i] = u
    vershv[i] = v
        
print(derevo(t,n,vershu,vershv))        
```

```python
def shooter(t, p1, p2):
    
    res = []
    for i in range(t):
        a0, prevb = p1[i] #–∫–ª–∞–¥—É—Ç—Å—è –¥–∞–Ω–Ω—ã–µ –ø–µ—Ä–≤–æ–≥–æ –º–æ–Ω—Å—Ç—Ä–∞, –µ–≥–æ –∑–¥–æ—Ä–æ–≤—å–µ –∏ –∫–∞–∫–æ–π –æ–Ω –Ω–∞–Ω–æ—Å–∏—Ç —É—Ä–æ–Ω –æ—Ç –≤–∑—Ä—ã–≤–∞
        ans, mn = 0, a0
        for p in p2[i]:
            a, b = p
            ans += max(0, a - prevb) #—É–±—å–µ—Ç –ª–∏ –≤–∑—Ä—ã–≤–æ–º —Å–ª–µ–¥—É—é—â–µ–≥–æ –º–æ–Ω—Å—Ç—Ä–∞, –µ—Å–ª–∏ –¥–∞, —Ç–æ 0, –µ—Å–ª–∏ –Ω–µ—Ç, —Ç–æ –æ—Å—Ç–∞—Ç–æ–∫ –∑–¥–æ—Ä–æ–≤—å—è
            mn = min(mn, a, prevb) #–≤—ã–±–∏—Ä–∞–µ—Ç—Å—è –º–∏–Ω–∏–º—É–º –º–µ–∂–¥—É –∑–¥–æ—Ä–æ–≤—å–µ–º –ø–µ—Ä–≤–æ–≥–æ –º–æ–Ω—Å—Ç—Ä–∞, –∑–¥–æ—Ä–æ–≤—å–µ–º –≤—Ç–æ—Ä–æ–≥–æ –∏ —É—Ä–æ–Ω–æ–º –æ—Ç –≤–∑—Ä—ã–≤–∞ –ø–µ—Ä–≤–æ–≥–æ (–¥–∞–ª–µ–µ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ)
            prevb = b #–ø—Ä–µ–¥—ã–¥—É—â–∏–º –∑–Ω–∞—á–µ–Ω–∏–µ–º —É—Ä–æ–Ω–∞ –æ—Ç –≤–∑—Ä—ã–≤–∞ —Ç–µ–ø–µ—Ä—å —Å—á–∏—Ç–∞–µ—Ç—Å—è —Å–ª–µ–¥—É—é—â–∏–π –º–æ–Ω—Å—Ç—Ä
        ans += max(0, a0 - prevb) #—É–º—Ä–µ—Ç –ª–∏ –º–æ–Ω—Å—Ç—Ä –∑–∞–¥–µ—Ç—ã–π –≤–∑—Ä—ã–≤–æ–º, –æ—Ç —ç—Ç–æ–≥–æ –∑–∞–≤–∏—Å–∏—Ç –Ω—É–∂–Ω–∞ –ª–∏ –µ—â—ë –ø—É–ª—è –∏ —Å–∫–æ–ª—å–∫–æ
        res.append(ans + min(mn, prevb))
        
    return res

t = int(input('–ö–æ–ª-–≤–æ –Ω–∞–±–æ—Ä–æ–≤: '))
p1 = []
p2 = []
for j in range(t):
    n = int(input('–ö–æ–ª-–≤–æ –º–æ–Ω—Å—Ç—Ä–æ–≤: '))
    p1.append(list(map(int, input().split()))) #–∫–ª–∞–¥–µ–º —Å—é–¥–∞ a0 –∏ prevb
    p2.append([list(map(int, input().split())) for i in range(n-1)]) #–∫–ª–∞–¥–µ–º —Å—é–¥–∞ a –∏ b
    
res = shooter(t, p1, p2)
print(*res)
```

–í —Ç—Ä–µ—Ç—å–µ–º —Ç–µ—Å—Ç–æ–≤–æ–º —Å–ª—É—á–∞–µ –∫–∞–∂–¥—ã–π —á–µ—Ç–≤–µ—Ä—Ç—ã–π –≥–æ—Å—Ç—å –ø–µ—Ä–µ–º–µ—â–∞–µ—Ç—Å—è –≤ –∫–æ–º–Ω–∞—Ç—É —Å –Ω–æ–º–µ—Ä–æ–º, –Ω–∞ 1 –±–æ–ª—å—à–∏–º –∏—Å—Ö–æ–¥–Ω–æ–≥–æ, –∞ –æ—Å—Ç–∞–ª—å–Ω—ã–µ –≥–æ—Å—Ç–∏ –ø–µ—Ä–µ–º–µ—â–∞—é—Ç—Å—è –≤ –∫–æ–º–Ω–∞—Ç—É —Å –Ω–æ–º–µ—Ä–æ–º, –Ω–∞ 5 –±–æ–ª—å—à–∏–º. –ú–æ–∂–Ω–æ –ø–æ–∫–∞–∑–∞—Ç—å, —á—Ç–æ –ø–æ-–ø—Ä–µ–∂–Ω–µ–º—É –≤ –∫–∞–∂–¥–æ–π –∫–æ–º–Ω–∞—Ç–µ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è —Ä–æ–≤–Ω–æ –æ–¥–∏–Ω –≥–æ—Å—Ç—å.

–í —á–µ—Ç–≤–µ—Ä—Ç–æ–º —Ç–µ—Å—Ç–æ–≤–æ–º —Å–ª—É—á–∞–µ –≥–æ—Å—Ç–∏, –∏—Å—Ö–æ–¥–Ω–æ –Ω–∞—Ö–æ–¥—è—â–∏–µ—Å—è –≤ –∫–æ–º–Ω–∞—Ç–∞—Ö 0 –∏ 1, –ø–µ—Ä–µ–º–µ—â–∞—é—Ç—Å—è –≤ –∫–æ–º–Ω–∞—Ç—É —Å –Ω–æ–º–µ—Ä–æ–º 3.

–í –ø—è—Ç–æ–º —Ç–µ—Å—Ç–æ–≤–æ–º —Å–ª—É—á–∞–µ –≥–æ—Å—Ç–∏, –Ω–∞—Ö–æ–¥—è—â–∏–µ—Å—è –≤ –∫–æ–º–Ω–∞—Ç–∞—Ö 1 –∏ 2, –ø–µ—Ä–µ–º–µ—â–∞—é—Ç—Å—è –≤ –∫–æ–º–Ω–∞—Ç—É —Å –Ω–æ–º–µ—Ä–æ–º 2.

```python
def chet(A, n, t):
    
    for i in range(t):  
        
        count = [0]*n
        
        for k in range(n): #—Ñ–æ—Ä–º—É–ª–∞ –∏–∑ —É—Å–ª–æ–≤–∏—è –∑–∞–¥–∞—á–∏
            
            if k % n < 0 or k % n > n-1: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
                raise Exception('–í—Å–µ–≥–¥–∞ –¥–æ–ª–∂–Ω–æ –≤—ã–ø–æ–ª–Ω—è—Ç—å—Å—è —É—Å–ª–æ–≤–∏–µ 0 ‚â§ ùëòmodùëõ ‚â§ ùëõ‚àí1')
    
            count[(A[k] + k) % n] += 1
            
        if not all(count): #–ø—Ä–æ–≤–µ—Ä—è–µ—Ç –∫–∞–∂–¥—ã–π —ç–ª–µ–º–µ–Ω—Ç –Ω–∞ –∏—Å—Ç–∏–Ω–Ω–æ—Å—Ç—å, –≥–¥–µ 1 - –∏—Å—Ç–∏–Ω–∞, 0 - –ª–æ–∂—å
            return 'NO'
        else:
            return 'YES'
        
t = int(input('–ö–æ–ª-–≤–æ —Ç–µ—Å—Ç–æ–≤—ã—Ö —Å–ª—É—á–∞–µ–≤: '))
sum = 0
result = []

if t < 1 or t > 10**4: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
    raise Exception('t –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å >= 1 –∏ t <= 10**4')
    
for i in range(t):
    n = int(input('–î–ª–∏–Ω–∞ –º–∞—Å—Å–∏–≤–∞: '))
    sum += n
    
    if sum > 2*10**5: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
        raise Exception('–°—É–º–º–∞ n –ø–æ –≤—Å–µ–º —Ç–µ—Å—Ç–æ–≤—ã–º —Å–ª—É—á–∞—è–º –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç 2*10^5!')
    if n < 1 or n > 10**4: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
        raise Exception('n –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å >= 1 –∏ n <= 10**5')
    
    A = [int(x) for x in input('–ú–∞—Å—Å–∏–≤ –∏–∑ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª: ').split()] #–º–∞—Å—Å–∏–≤ –∏–∑ —Ü–µ–ª—ã—Ö —á–∏—Å–µ–ª –¥–ª—è –ø–µ—Ä–µ–Ω–æ—Å–æ–≤ –≥–æ—Å—Ç–µ–π
    
    for i in A:
        if i < -10**9 or i > 10**9: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
            raise Exception('a –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å >= -10**9 –∏ a <= 10**9')
            
    result.append(chet(A, n, t))
    
print(*result)
```

```python
def nok(n): #–≤—ã—à–ª–∞ –∑–∞–∫–æ–Ω–æ–º–µ—Ä–Ω–æ—Å—Ç—å, —á—Ç–æ –Ω–∞–∏–º–µ–Ω—å—à–µ–µ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ, –∫—Ä–∞—Ç–Ω–æ–µ —Ç—Ä–µ–º —á–∏—Å–ª–∞–º –º–µ–Ω—å—à–∏–º n - —ç—Ç–æ 
    if n <= 2:
        return(n)
    elif n % 2!= 0:
        return n*(n - 1)*(n - 2)
    elif n % 3!= 0:
        return n*(n - 1)*(n - 3)
    elif n % 2 == 0:
        return (n - 1)*(n - 2)*(n - 3)
    
n=int(input('–í–≤–µ–¥–∏—Ç–µ n: '))

if n < 1 or n > 10**6: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
    raise Exception('n –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å >= 1 –∏ t <= 10**6')
    
print(nok(n))
```

```python
from math import *

def lcm(x,y): #–Ω–∞–∏–º–µ–Ω—å—à–µ–µ –æ–±—â–µ–µ –∫—Ä–∞—Ç–Ω–æ–µ
    
    return x*y//gcd(x,y) #gsd - –Ω–∞–∏–±–æ–ª—å—à–∏–π –æ–±—â–∏–π –¥–µ–ª–∏—Ç–µ–ª—å (–∏–∑ math)

n = int(input('–í–≤–µ–¥–∏—Ç–µ —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ:'))

if n < 2 or n > 10**5: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
    raise Exception('n –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å >= 2 –∏ n <= 10**5')
    
l = list(map(int,input('–í–≤–µ–¥–∏—Ç–µ n —á–∏—Å–µ–ª a: ').split()))

e = 0  #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
for i in l:
    e += 1
    if i < 1 or n > 2*10**5: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
        raise Exception('a –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å >= 2 –∏ n <= 2*10**5')
if e != n:
    raise Exception('–ö–æ–ª-–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ a –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å —Ä–∞–≤–Ω—ã–º n!')

#---------------------------------------------------------------------------

t = [0]*(n+1) #—Å–æ–∑–¥–∞—é –º–∞—Å—Å–∏–≤ –∏–∑ –Ω—É–ª–µ–π
t[-1] = l[-1] #–ø–æ—Å–ª–µ–¥–Ω–∏–π —ç–ª–µ–º–µ–Ω—Ç l —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è —ç–ª–µ–º–µ–Ω—Ç–æ–º –ø–æ—Å–ª–µ–¥–Ω–∏–º –¥–ª—è t

for i in range(len(l)-1,0,-1):
    
    t[i] = gcd(l[i],t[i+1]) #–Ω–∞—Ö–æ–∂—É –Ω–∞–∏–±–æ–ª—å—à–∏–π –æ–±—â–∏–π –¥–µ–ª–∏—Ç–µ–ª—å –º–µ–∂–¥—É —ç–ª–µ–º–µ–Ω—Ç–æ–º a –∏ —ç–ª–µ–º–µ–Ω—Ç–æ–º t[i+1], –≤—Å—ë –∫–ª–∞–¥–µ—Ç—Å—è –≤ –ø—Ä–µ–¥—ã–¥—É—â–∏–π –∏–Ω–¥–µ–∫—Å t    
                            #—Ç–µ–ø–µ—Ä—å –≤ t –ª–µ–∂–∞—Ç –Ω–∞–∏–± –æ–±—â –¥–µ–ª–∏—Ç–µ–ª–∏
b=[0]*n

for i in range(1,n-1): #–∫–ª–∞–¥—É –≤ b (–∞ * –Ω–∞–∏–± –æ–±—â –¥–µ–ª–∏—Ç–µ–ª—è –∏–∑ t)//–ø—Ä–µ–¥—ã–¥—É—â–µ–µ —á–∏—Å–ª–æ –≤ t, –ø–æ–ª—É—á–∞–µ—Ç—Å—è, —á—Ç–æ –≤ b –ª–µ–∂–∞—Ç –∫—Ä–∞—Ç–Ω—ã–µ —á–∏—Å–ª–∞
    
    b[i]=(l[i]*t[i+1])//t[i]
    
x=b[0]

for i in range(1,n-1):
    
    x=gcd(x,b[i]) #–Ω–∞–∏–±–æ–ª—å—à–∏–π –æ–±—â–∏–π –¥–µ–ª–∏—Ç–µ–ª—å —Å—Ä–µ–¥–∏ –º–∞—Å—Å–∏–≤–∞ —Å –∫—Ä–∞—Ç–Ω—ã–º–∏
    
if len(l) == 1:
    
    print(l[0])
    
elif len(l) == 2:
    
    print(lcm(l[0],l[1]))
    
else:
    
    print(x)
```

```python
def riddle(n, C):
    
    prev = ''
    c = cs = 0
    for i in range(n):
        x = y = 10**15 #–ø—Ä–∏—Å–≤–∞–∏–≤–∞—é –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –±–æ–ª—å—à–æ–µ —á–∏—Å–ª–æ, —á—Ç–æ–±—ã –∏—Å–∫–∞—Ç—å –º–∏–Ω–∏–º—É–º
        if arrs[i] >= prev: x = c #—Å—Ä–∞–≤–∏–Ω–≤–∞—é –≤–µ—Å –¥–≤—É—Ö —Å—Ç—Ä–æ–∫
        if arrs[i] >= prev[::-1]: x = min(x,cs) #—Å—Ä–∞–≤–Ω–∏–≤–∞—é –≤–µ—Å –ø–µ—Ä–≤–æ–π –∏ –ø–µ—Ä–µ–≤–µ—Ä–Ω—É—Ç–æ–π –≤—Ç–æ—Ä–æ–π
        if arrs[i][::-1] >= prev: y = c + C[i] #—Å—Ä–∞–≤–Ω–∏–≤–∞—é –≤–µ—Å –ø–µ—Ä–µ–≤–µ—Ä–Ω—É—Ç–æ–π –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π
        if arrs[i][::-1] >= prev[::-1]: y = min(y,cs+C[i])
        c = x
        cs = y
        prev = arrs[i]
    ans = min(c,cs)
    if ans == 10**15: ans = -1
        
    return ans

n = int(input('–ö–æ–ª-–≤–æ —Å—Ç—Ä–æ–∫: '))

if n < 2 or n > 10**5: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
    raise Exception('n –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å >= 2 –∏ n <= 10**5')
    
C = list(map(int, input('–≠–Ω–µ—Ä–≥–æ–∑–∞—Ç—Ä–∞—Ç–Ω–æ—Å—Ç—å –¥–µ–π—Å—Ç–≤–∏–π: ').split()))
if n != len(C):
    raise Exception('–ù–µ–≤–µ—Ä–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–∏—Å–µ–ª!')
for i in C:
    if C < 0 or C > 10**9: #–∑–∞—â–∏—Ç–∞ –æ—Ç –¥—É—Ä–∞–∫–∞
        raise Exception('ci –¥–æ–ª–∂–Ω–æ –±—ã—Ç—å >= 0 –∏ ci <= 10**9')

for i in range(n):
    arrs.append(str(input()))
    
print('–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ–π —ç–Ω–µ—Ä–≥–∏–∏: ', riddle(n, C))
```
